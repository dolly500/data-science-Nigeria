```python
x = re.drop('subscribed',axis=1)
y = re['subscribed']
```

```python
x_train, x_test ,y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=45)
```

```python
lr = LogisticRegression()
rf = RandomForestClassifier(n_estimators=100, random_state=45)
xg = XGBClassifier(base_score=0.5, booster='gbtree', n_estimators=200, max_depth=3, min_child_weight=0.9, 
                   eta=0.2, gamma=2)
gb = GradientBoostingClassifier(max_depth=4, subsample=0.8, n_estimators=100)
lg = LGBMClassifier(max_depth=5)
ct= CatBoostClassifier(depth=5)
Ad = AdaBoostClassifier()
cm = DecisionTreeClassifier()
cm = confusion_matrix
ds = cross_val_score
```

```python
lg.fit(x_train,y_train)
pred = lg.predict(x_test)
score = accuracy_score(y_test,pred)
print('f1 score is', accuracy_score)
 
```

```python
from tqdm import tqdm
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=23,shuffle=True,random_state=42)
oof_preds=np.zeros(x.shape[0])
test_preds=[]
model=gb
for train_indices,test_indices in tqdm(skf.split(x,y)):
    train_x,test_x=x.iloc[train_indices],x.iloc[test_indices]
    train_y,test_y=y.iloc[train_indices],y.iloc[test_indices]
    model.fit(train_x,train_y)
    pred = model.predict(test_x)
    oof_preds[test_indices]=pred
    score = accuracy_score(test_y,pred)
    print('accuracy_score is', score)
    test_pred=model.predict(rd)
    test_preds.append(test_pred)
oof_score=accuracy_score(y,oof_preds)
print('out_of_fold_score is', oof_score)
predictions=np.mean(test_preds,axis=0)
predictions=[1 if c>0.5 else 0 for c in predictions]
```
